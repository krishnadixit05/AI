```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load the dataset
df = pd.read_csv('imdb_top_1000.csv')
```

```python
df
```

```python
if 'Certificate' in df.columns:
    df_tv = df[df['Certificate'] == 'TV Series'].copy()
    df_tv = df[df['Certificate'] == 'TV Series'].copy()
    # Fill missing values for df_tv
    for feature in ['Genre', 'Director', 'Star1', 'Star2', 'Star3', 'Star4', 'Overview']:
        df_tv[feature] = df_tv[feature].fillna('')
else:
    print(f"Column 'Certificate' not found. Available columns: {df.columns.tolist()}")
```

```python
# Combine features into a single string
def combine_features(row):
    return f"{row['Genre']} {row['Director']} {row['Star1']} {row['Star2']} {row['Star3']} {row['Star4']} {row['Overview']}"

# Apply the function to create the 'combined_features' column
if 'df_tv' in locals(): #This makes sure that df_tv got created before trying to use it.
    df_tv['combined_features'] = df_tv.apply(combine_features, axis=1)
else:
    print("df_tv was not created, please check the previous cells and ensure they have been executed")
```

```python
# Vectorize the combined features
vectorizer = TfidfVectorizer(stop_words='english')
feature_vectors = vectorizer.fit_transform(df_tv['combined_features'])

# Compute cosine similarity
similarity = cosine_similarity(feature_vectors)

# Function to get recommendations
def recommend_series(title, similarity=similarity):
    if title not in df_tv['Series_Title'].values:
        print(f"Series '{title}' not found in the dataset.")
        return []
    idx = df_tv[df_tv['Series_Title'] == title].index[0]
    sim_scores = list(enumerate(similarity[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Exclude the series itself
    series_indices = [i[0] for i in sim_scores]
    return df_tv['Series_Title'].iloc[series_indices]

# Example usage
recommended = recommend_series('Breaking Bad')
print("Top 10 recommendations for 'Breaking Bad':")
print(recommended)
```
